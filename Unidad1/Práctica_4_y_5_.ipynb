```python
import random
edades =[]
for i in range(100):
    numero_aleatorio= random.randint(18,70)
    edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(edades, columns=['edad'])
df.head()
```

```python
tabla_frecuencias= df['edad'].value_counts(). reset_index()
tabla_frecuencias.columns = ['edad', 'frecuencia' ]
tabla_frecuencias = tabla_frecuencias.sort_values(by='edad')

print("tabla de distribución de frecuencia")
tabla_frecuencias
```

```python
tabla_frecuencias['frec relativa']= tabla_frecuencias['frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['frec porcentual']= tabla_frecuencias['frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['frec acumulada']= tabla_frecuencias['frecuencia'].cumsum()
tabla_frecuencias
```

# regla de sturges
## se puede calcular en base a una fórmula, tomando el tamaño de la muestra
C= 1+log base 2 (N)
Dónde N es el tamaño de la muestra

```python
import math
```

```python
print (edades)
```

```python
df= pd.DataFrame({'edad': edades})
df
```

```python
n = len(edades)
k= int(1+ math.log2(n))
print("valor de n= ", n , "\n Número de intervalos= ",k)
```

```python
tabla_frecuencia= pd.cut(df['edad'] , bins = k, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns=['clase','frec']
tabla_frecuencia
```

```python
tabla_frecuencia=tabla_frecuencia.sort_values(by='clase')
tabla_frecuencia
```

```python
tabla_frecuencia['frec rel'] = tabla_frecuencia['frec']/n
tabla_frecuencia['frec porc'] = tabla_frecuencia['frec'] /n *100
tabla_frecuencia['frec Ac'] = tabla_frecuencia['frec'].cumsum()
tabla_frecuencia['frec rel Ac'] = tabla_frecuencia['frec rel'].cumsum
tabla_frecuencia['frec porc Ac'] = tabla_frecuencia['frec porc'].cumsum()
print(tabla_frecuencia)
```

# Práctica 5 diagramas de cajas y bigotes

### las líneas se usan para indicar la variabilidad fuera de los cuarteles superior e inferior
###la caja representa donde está la mayor cantidad de datos
### la línea dentro de la caja representa el valor de la media

```python
edades2 =[]
for i in range(100):
    numero_aleatorio= random.randint(18,70)
    edades2.append(numero_aleatorio)
print(edades2)
```

```python
fig,(ax1,ax2)= plt.subplots(nrows=1,ncols=2, figsize=(9,4))
bplot1=ax1.boxplot(edades,
                   vert=True, # alineamiento vertical de caja
                   patch_artist=True)
ax1.set_title('edades encuesta 1')

bplot2=ax2.boxplot(edades2,
                   vert=True, # alineamiento vertical de caja
                   patch_artist=True)
ax2.set_title('edades encuesta 2')

plt.show()
```
