# Permutaciones con python
### 1. existen 10 asistentes de profesor disponibles para calificar examenes en un curso de calculo en una gran universidad.
### 2. el primer examen se compone de 4 preguntas y el profesor desea seleccionar un asistente diferente para calificar cada pregunta(solo un asistente por pregunta)

### 3. de cuantas maneras se pueden elegir los asistentes para calificar?

### 4. en este caso n = tamaño del grupo = 10 y k = tamaño del subconjunto = 4
### 5. el numero de permutaciones es:


image.png

```python
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes = ["a", "b","c","d","e","f","g","h","i","j"]
n = len(asistentes)
k = 4

Pr= mt.factorial(n)/mt.factorial(n-k)
print("existen ", Pr , " posibles permutaciones sin repeticion del total de ",
      n , " asistentes " , " en grupos de ", k , " en ", k)
print("aqui las permutaciones")
permutaciones_asistentes = list(permutations(asistentes,k))
for permutation in permutaciones_asistentes:
    print(permutation)

```
