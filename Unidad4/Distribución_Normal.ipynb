#Ejercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media= 70
desv_std= 4
num_datos = 500

datos = np.random.normal(media,desv_std,num_datos)
datos[0:50]
```

```python
x= np.linspace(media-3 * desv_std,media+3*desv_std,num_datos)

pdf = 1/(desv_std*np.sqrt(2* np.pi))*np.exp(-(x-media)**2/(2*desv_std**2))

print("imprimir los primeros 50 valores de x",x[0:50]," , y los primeros 50 pdf",pdf[0:50])
```

```python
# Graficar histograma

plt.figure(figsize=(6, 4))

plt.hist(datos, bins=15, alpha=0.4, color='b')

plt.title('Distribuci√≥n normal')

plt.xlabel('Peso en kg de los estudiantes')

plt.grid(True)

plt.show()
```

```python

# Visualizar el PDF

plt.figure(figsize=(6, 4))

plt.plot(x, pdf, label='Normal: (media: {}; ds: {})'.format(np.round(media, 2), np.round(desv_std, 2)))

plt.title('Funci√≥n de densidad de probabilidad (PDF) Distribuci√≥n normal')

plt.xlabel('Pesos en kg de los estudiantes')

plt.ylabel('Densidad de probabilidad')

plt.legend()

plt.grid(True)

plt.show()
```

```python
# Graficar histograma

plt.figure(figsize=(6, 4))

plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')

plt.plot(x, pdf, 'k', linewidth=2, )

plt.title('Distribuci√≥n normal')

plt.xlabel('Peso en kg de los estudiantes')

plt.grid(True)

plt.show()
```

#Ejercicio 2

```python
plt.figure(figsize=(6,4))

# Par√°metros de la distribuci√≥n normal

media = 70 # Media
desv_std = 4 # Desviaci√≥n est√°ndar
total_estudiantes = 500

# VARIABLE para calcular la probabilidad de que un estudiante pese menos de 75 kg

peso_limite=  79

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la funci√≥n de distribuci√≥n acumulativa (CDF) de la distribuci√≥n normal

prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
# C√°lculo del n√∫mero de estudiantes que pesan menos de 75 kg

estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg
print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("N√∫mero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

# Graficar la distribuci√≥n normal

x = np.linspace(media-  4* desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
# Cu√°ntos estudiantes pesan menos de 62 kg.

plt.figure(figsize=(6,4))

# Par√°metros de la distribuci√≥n normal

media = 70 # Media
desv_std = 4 # Desviaci√≥n est√°ndar
total_estudiantes = 500

# Variable para calcular la probabilidad de que un estudiante pese menos de 75 kg

peso_l√≠mite = 62

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos

# de 75 kg utilizando la funci√≥n de distribuci√≥n acumulativa (CDF) de la distribuci√≥n normal

prob_menos_62kg = norm.cdf(peso_l√≠mite, media, desv_std)

# C√°lculo del n√∫mero de estudiantes que pesan menos de 75 kg

estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("N√∫mero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_62kg)
```

```python
# Calcular la distribuci√≥n normal

x = np.linspace(media- 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribuci√≥n normal

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg, \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
#¬øCu√°ntos estudiantes pesan m√°s de 79 kg?

plt.figure(figsize=(6,4))

# Par√°metros de la distribuci√≥n normal

media = 70 # Media
desv_std = 4 # Desviaci√≥n est√°ndar
peso_limite = 79 # Peso l√≠mite
total_estudiantes = 500

# Calculamos la probabilidad de que un estudiante pese menos o igual a 79 kg

prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese m√°s de 79 kg es complementaria

prob_mas_79kg = 1 - prob_menos_79kg

# Calculamos el n√∫mero aproximado de estudiantes que pesan m√°s de 79 kg

estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg
print("Probabilidad de que un estudiante pese m√°s de 79 kg:", prob_mas_79kg)
print("N√∫mero aproximado de estudiantes que pesan m√°s de 79 kg:", estudiantes_mas_79kg)
```

```python
# Calcular la funci√≥n de densidad de probabilidad de la distribuci√≥n normal

x = np.linspace(media -4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese m√°s de 79 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
